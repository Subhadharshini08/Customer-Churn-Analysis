```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
```

```python
df=pd.read_csv("Telco_Customer_Cleaned.csv")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Information about the dataset
```

```python
print("Dataset Info:")
print(df.info())
print("\n Basic Statistics:")
print(df.describe())
```

```python
# Churn Distribution
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Churn', data=df, hue='Churn', palette=['#2ecc71', '#e74c3c'], legend=False)
plt.title('Customer Churn Distribution', fontsize=16)
plt.xlabel('Churn', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Add data labels
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', 
                       (p.get_x() + p.get_width()/2, p.get_height()),
                       ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Churn by Contract Type
```

```python
# Churn by Contract Type
plt.figure(figsize=(8, 5))
ax=sns.countplot(x='Contract', data=df, hue='Churn', 
              palette=['#2ecc71', '#e74c3c'], legend=True)
plt.title('Churn by Contract Type', fontsize=16)
plt.xlabel('Contract Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
# Add data labels
for p in ax.patches:
    if p.get_height()>0:
        ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Churn by Internet Service
```

```python
plt.figure(figsize=(8, 5))
ax=sns.countplot(x='InternetService', data=df, hue='Churn', 
              palette=['#2ecc71', '#e74c3c'], legend=True)
plt.title('Churn by Internet Service', fontsize=16)
plt.xlabel('Internet Service', fontsize=12)
plt.ylabel('Count', fontsize=12)
# Add data labels
for p in ax.patches:
    if p.get_height() > 0:  # only show label if height > 0
        ax.annotate(f'{int(p.get_height())}',
                    (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Churn by Payment Method
```

```python
plt.figure(figsize=(8, 5))
ax=sns.countplot(x='PaymentMethod', data=df, hue='Churn', 
              palette=['#2ecc71', '#e74c3c'], legend=True)
plt.title('Churn by Payment Method', fontsize=16)
plt.xlabel('Payment Method', fontsize=12)
plt.ylabel('Count', fontsize=12)
# Add data labels
for p in ax.patches:
    if p.get_height() > 0:  # only show label if height > 0
        ax.annotate(f'{int(p.get_height())}',
                    (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Churn by Senior Citizen
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].replace({0: 'Non-Senior', 1: 'Senior'})
plt.figure(figsize=(8, 5))
ax=sns.countplot(x='SeniorCitizen', data=df, hue='Churn', 
              palette=['#2ecc71', '#e74c3c'], legend=True)
plt.title('Churn by Senior Citizen', fontsize=16)
plt.xlabel('Senior Citizen', fontsize=12)
plt.ylabel('Count', fontsize=12)
# Add data labels
for p in ax.patches:
    if p.get_height() > 0:  # only show label if height > 0
        ax.annotate(f'{int(p.get_height())}',
                    (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Monthly Charges Distribution by Churn
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='MonthlyCharges', hue='Churn',
             palette=['#2ecc71', '#e74c3c'], bins=30)
plt.title('Monthly Charges Distribution by Churn', fontsize=16)
plt.xlabel('Monthly Charges', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.show()
```
